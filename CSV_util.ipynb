```python
# imports
from geopy.geocoders import Nominatim
import pandas as pd
import geopandas
from shapely.geometry import Point, Polygon
```

```python
data = pd.read_csv('2013_Events.csv')
```

```python
result = [x + ', ' + y for x, y in zip(data['Oblast'], data['Location'])]
data.insert(5, 'Location String', result)
```

```python
coord_dict = {}
```

```python
%store -r coord_dict
```

```python
geolocator = Nominatim(user_agent="Your_Name")

def get_coords(name):
    print('Getting coords for: ' + name)
    if name in coord_dict:
        #print(name + ' is in the dict')
        return coord_dict[name]
    else:
        coords = geolocator.geocode(name)
        if coords:
            coord_dict[name] = Point([coords.longitude, coords.latitude])
            return coord_dict[name]
        else:
            return None
```

```python
result2 = [get_coords(x) for x in data['Location String']]
```

```python
%store coord_dict
```

```python
data.insert(6, 'coordinates', result2)
```

```python
data = geopandas.GeoDataFrame(data, geometry='coordinates')
```

```python
# remove entries for which no location could be found
data = data.loc[data['coordinates'] != None]
```

```python
data.head()

xcoords = [point.x for point in data['coordinates']]
ycoords = [point.y for point in data['coordinates']]

data.insert(5, 'x', xcoords)
data.insert(6, 'y', ycoords)

data.head()
```

```python
def checkDate(str):
    try:
        return pd.to_datetime(str)
    except AttributeError:
        return 0
```

```python
# add month column, manually adjust scaling per year
months = [48 + x.month if x !=0 else 0 for x in checkDate(data["Event start date"])]
data.insert(7, 'month', months)
```
